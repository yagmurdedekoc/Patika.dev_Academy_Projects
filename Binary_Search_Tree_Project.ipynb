```python
# Binary Search Tree Stages

class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def insert_into_bst(root, value):
    if root is None:
        return TreeNode(value)
    if value < root.value:
        root.left = insert_into_bst(root.left, value)
    else:
        root.right = insert_into_bst(root.right, value)
    return root

def add_edges_to_graph(root, graph, pos, x=0, y=0, layer=1):
    if root is not None:
        graph.add_node(root.value, pos=(x, y))
        if root.left:
            graph.add_edge(root.value, root.left.value)
            add_edges_to_graph(root.left, graph, pos, x - 1 / layer, y - 1, layer + 1)
        if root.right:
            graph.add_edge(root.value, root.right.value)
            add_edges_to_graph(root.right, graph, pos, x + 1 / layer, y - 1, layer + 1)

def draw_bst(root):
    graph = nx.DiGraph()
    pos = {}
    add_edges_to_graph(root, graph, pos)
    pos = nx.get_node_attributes(graph, 'pos')
    
    plt.figure(figsize=(10, 7))
    nx.draw(graph, pos, with_labels=True, node_size=2000, node_color="lightblue", font_size=10, font_weight="bold", arrows=False)
    plt.show()

# Binary Search Tree -> [7, 5, 1, 8, 3, 6, 0, 9, 4, 2]

data = [7, 5, 1, 8, 3, 6, 0, 9, 4, 2]

root = None
for value in data:
    root = insert_into_bst(root, value)

draw_bst(root)
```

```python
# Binary Search Tree Stages

# Binary Search Tree is a tree structure in which the values of the nodes in the left subtree of each node are smaller than itself and the values of the nodes in the right subtree are larger than itself.
# Given sequence: [7,5,1,8,3,6,0,9,4,2] 
# We select the first element as the root, then add the others in order.

# 1. Root: 7
# The first element of the array, 7, is selected as the root node.

# 2. Adding: 5
# Since 5 is less than 7, it is placed to the left of 7

# Tree state:

#      7
#     /
#    5

# 3. Adding: 1
# Since 1 is less than 7 and less than 5, it is placed to the left of 5.

# Tree state:

#      7
#     /
#    5
#   /
#  1

# 4. Adding: 8
# Since 8 is greater than 7, it is placed to the right of 7.

# Tree state:

#      7
#     / \
#    5   8
#   /
#  1

# 5. Adding: 3
# Since 3 is less than 7 and less than 5 but greater than 1, it is placed to the right of 1.

# Tree state:

#      7
#     / \
#    5   8
#   /
#  1
#   \
#    3

# 6. Adding: 6
# Since 6 is less than 7 and greater than 5, it is placed to the right of 5.

# Tree state:

#      7
#     / \
#    5   8
#   / \
#  1   6
#   \
#    3

# 7. Adding: 0
# Since 0 is less than 7, less than 5 and less than 1, it is placed to the left of 1.

# Tree state:

#         7
#        / \
#       5   8
#      / \   
#     1   6   
#    / \
#   0   3

# 8. Adding: 9
# Since 9 is greater than 7 and greater than 8, it is placed to the right of 8.

# Tree state:

#         7
#        / \
#       5   8
#      / \   \
#     1   6   9
#    / \
#   0   3

# 9. Adding: 4
# Since 4 is less than 7, less than 5, greater than 1, greater than 3, it is placed to the right of 3.

#         7
#        / \
#       5   8
#      / \   \
#     1   6   9
#    / \
#   0   3
#        \
#         4

# 10. Adding: 2
# Since 2 is less than 7, less than 5, greater than 1 and less than 3, it is placed to the left of 3.

#         7
#        / \
#       5   8
#      / \   \
#     1   6   9
#    / \
#   0   3
#      / \
#     2   4

# Conclusion
# When the elements of the array are added to the tree structure, the Binary Search Tree is as above. In the tree, the values to the left of each node are placed in such a way that the values to the left of each node are smaller than itself and the values to the right of each node are larger than itself.
```
